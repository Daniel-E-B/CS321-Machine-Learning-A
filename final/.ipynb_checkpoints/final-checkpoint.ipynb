```python
data = pd.read_csv('partially_filtered_appdata.csv', index_col=0)
```

```python
# onehot encoding genre
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
onehot_genres = onehotencoder.fit_transform(data["genre"].values.reshape(-1, 1)).toarray()
data = data.drop("genre", 1)
genres_df = pd.DataFrame(onehot_genres)
genres_df.columns = onehotencoder.get_feature_names()
data = pd.concat([data, genres_df], axis=1)
data.head()
```
